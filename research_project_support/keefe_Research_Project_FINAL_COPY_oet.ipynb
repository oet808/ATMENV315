```python
i=0
firstlist=[]
for value in a:
    first=float(a[i])
    print(first)
    firstlist.append(first)
    i=i+1
    
firstarray=np.array(firstlist)
```

```python
s1=np.mean(firstarray)
print(s1)
```

```python
s2=np.std(firstarray)
print(s2)
```

```python
firstarray
mean1=np.mean(firstarray)
std1=np.std(firstarray)
size1=np.size(firstarray)
df=size1-1
stderr1=std1/np.sqrt(size1)
q1=t.ppf(0.025, df)
q2=t.ppf(0.975, df)
ci1=mean1+stderr1*q1
ci2=mean1+stderr1*q2
print('lower = ' + str(ci1) + " upper = " + str(ci2))
```
