```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename="epa_ozone_data_sample_nyc_2020.csv"

df=pd.read_csv(filename)

df
```

```python
df.columns
```

```python
df['Date'] # need to be converted into datetime numerical data
dfg=df.groupby('Site ID')
dfg.groups
```

```python
df1=dfg.get_group(361192004) # one station
x=df1['Daily Max 8-hour Ozone Concentration'].values
```
