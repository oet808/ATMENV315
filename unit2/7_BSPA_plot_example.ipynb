### Main challenges

- How to deal with the date and time information in the data table and make a plot?


````python
# code development from scratch
import pandas as pd
import matplotlib.pyplot as plt
````

````python
df=pd.read_csv("/nfs/home11/staff/timm/Public/Data/BSPA_HOBO_data.csv")
````

````python
time=df['Date_Time_GMT-04:00']
temp=df['Temp']
plt.plot(temp)
plt.ylabel('temperature [deg F]')
plt.show()
````

````python
# we cannot use the date and time data from the CSV table
# directly in the plot function.
# the strings have to be converted into numeric date and time data
# this is done here
````

````python
x=pd.to_datetime(time)
````

````python
plt.plot(x,temp,color='red',label='BSPA Hobo logger')
plt.legend()
plt.ylabel('temperature [deg F]')
plt.xlabel('date and time')
plt.xticks(rotation = 90)
plt.show()
````

This was a quick demonstration of the essential coding steps that allow us to a time series with data stored in a CSV file. The plotted time series can be improved with additial customization of the axis tick labels, but this is a bit tedious and takes the attention away from the essential steps used here.

See for example [https://www.earthdatascience.org](https://www.earthdatascience.org/courses/use-data-open-source-python/use-time-series-data-in-python/date-time-types-in-pandas-python/customize-dates-matplotlib-plots-python/) for customized plot appearances.

````python
p = bokeh.plotting.figure(
    frame_height=200,
    frame_width=500,
    toolbar_location='above',
    x_axis_label='time',
    y_axis_label='temperature [deg F]'
)

p.line(x,temp,color='red')
bokeh.io.show(p)
````
