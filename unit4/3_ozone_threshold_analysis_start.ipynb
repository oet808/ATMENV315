```python
path="/nfs/home11/staff/timm/Public/Data/hw2/"
city_id='phoenix_az'
pollutant='ozone'
datacolumn='Daily Max 8-hour Ozone Concentration'
filename=city_id+'_'+pollutant+'.csv'
# open the CVS file
df=pd.read_csv(path+filename)
# update the column with better numerical / plot-friendly datetime values
df['Date']=pd.to_datetime(df['Date'], format="%m/%d/%y")
# extract columns from dataframe for simpler use
day=df['Date']
concentration=df['Daily Max 8-hour Ozone Concentration']
fig,ax=plt.subplots(1,1,figsize=[10,4])

ax.plot(day,concentration,'black')
ax.axhline(0.07,color='red',linestyle='--')
ax.text(day[0],0.072,'threshold',size=14,color='red')
fig.show()
```
